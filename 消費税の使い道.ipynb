# 消費税と社会保障費の推移分析ノートブック
このNotebookでは、日本の消費税収と社会保障関係費の推移を比較し、「本当に消費税は社会保障に使われているのか？」を可視化と分析で学びます。

```python
# Step 1: データの読み込み
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('consumption_tax_social_security.csv')
df.head()
```

```python
# Step 2: データの可視化
import matplotlib.pyplot as plt
import matplotlib

# Windowsの日本語フォントを指定（MS Gothic など）
plt.rcParams['font.family'] = 'MS Gothic'

# 負の値（マイナス記号）が表示されない問題を防止
plt.rcParams['axes.unicode_minus'] = False


plt.figure(figsize=(12,6))
plt.plot(df['年度'], df['消費税収（兆円）'], label='消費税収', marker='o')
plt.plot(df['年度'], df['社会保障関係費（兆円）'], label='社会保障関係費', marker='o')
plt.title('消費税収と社会保障関係費の推移')
plt.xlabel('年度')
plt.ylabel('兆円')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 必要なライブラリのインポート
import pandas as pd
import matplotlib.pyplot as plt

# 日本語フォントの設定（Windows用）
plt.rcParams['font.family'] = 'MS Gothic'
plt.rcParams['axes.unicode_minus'] = False

# データ読み込み
df = pd.read_csv('consumption_tax_full_with_tax_breakdown.csv')
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['年度'], df['消費税収（兆円）'], label='消費税収', marker='o')
plt.plot(df['年度'], df['法人税収（兆円）'], label='法人税収', marker='o')
plt.plot(df['年度'], df['所得税収（兆円）'], label='所得税収', marker='o')
plt.title('主要税収の推移')
plt.xlabel('年度')
plt.ylabel('兆円')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['年度'], df['全税収（兆円）'], label='全税収（3税）', marker='o')
plt.plot(df['年度'], df['社会保障関係費（兆円）'], label='社会保障関係費', marker='o')
plt.title('全税収と社会保障費の比較')
plt.xlabel('年度')
plt.ylabel('兆円')
plt.legend()
plt.grid(True)
plt.show()
```

✅ ① 消費税の割合が税収全体に占める割合（%）
📈 グラフで見えること：

・年を追うごとに消費税の割合が増えている

・特に2014年（8%）と2019年（10%）の増税後に割合がぐっと上がる

・現在では、税収全体の中で約35〜40%を占める年もある

🧾 つまり…
消費税は「国民みんなから幅広く集める税」なので、税収の柱になっている
→ 逆に言うと、消費税の負担が家計にじわじわ効いているとも言える

✅ ② 法人税・所得税の推移（景気との関係）
📈 グラフで見えること：

・法人税収は景気に敏感で、2008年のリーマンショック後に急落

・所得税収は比較的安定しているが、横ばいに近い

・消費税のような右肩上がりの伸びはない

🧾 つまり…
法人税は「企業の利益」に連動するため、景気が悪化すると税収も落ちる

消費税は景気に関係なく取れるため、政府にとって安定的な財源になっている

✅ ③ 消費税率の変更年（1997年・2014年・2019年）と税収の変化
📈 グラフで見えること：
・1997年：3% → 5%

・2014年：5% → 8%

・2019年：8% → 10%

それぞれの年の直後に消費税収が急増している
→ ただし、景気の落ち込みで一時的に減る年もある（特に1997年）

🧾 つまり…
・増税＝税収アップにはなるが、景気や消費にブレーキがかかるリスクもある

・そのバランスをとるのが政策の難しさ
